```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline
import gradio as gr
import requests
from bs4 import BeautifulSoup
```

```python
def scrape_site(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    text = " ".join([p.get_text() for p in soup.find_all("p")])
    return text
```

```python
urls = [
    "https://doaj.org/?spm=a2ty_o01.29997173.0.0.6774c921AK3gx4",
    "https://openstax.org/?spm=a2ty_o01.29997173.0.0.6774c921AK3gx4"
    "https://en.wikibooks.org/?spm=a2ty_o01.29997173.0.0.6774c921AK3gx4"
]
```

```python
all_texts = []
for url in urls:
    try:
        all_texts.append(scrape_site(url))
    except:
        print(f"Error scraping {url}")
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = splitter.create_documents(all_texts)
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(docs, embeddings)
retriever = vectorstore.as_retriever()
```

```python
generator = pipeline("text2text-generation", model="facebook/bart-large-cnn")
llm = HuggingFacePipeline(pipeline=generator)
qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)
```

```python
def answer_question(question):
    try:
        return qa_chain.run(question)
    except Exception as e:
        return str(e)
```

```python
iface = gr.Interface(
    fn=answer_question,
    inputs=gr.Textbox(lines=2, placeholder="Type your question here... 🤔"),
    outputs="text",
    title="🌐 Web QA System",
    description="Ask any question about the selected websites and get answers from the extracted information! 📝"
)

iface.launch()
```
